```python
for i in df2["fname"]:
    path_wav = "./dataset/heartbeat-sounds/" + i
    sounds = AudioSegment.from_file(path_wav)
    data = np.array(sounds.get_array_of_samples())
    x = data[::sound.channels]
    plt.plot(x[::10])
    plt.grid()
    plt.show()
```

```python
import wave
import numpy as np
import matplotlib.pyplot as plt
```

```python
wf = wave.open(path_wav , "r" )
buf = wf.readframes(wf.getnframes())
```

```python
# バイナリデータを整数型（16bit）に変換
data = np.frombuffer(buf, dtype="int16")

# グラフ化
plt.plot(data)
plt.grid()
plt.show()
```
