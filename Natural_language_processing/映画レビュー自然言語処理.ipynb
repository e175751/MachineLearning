まずは、``MeCab``を使い単語を``品詞``別に分けていく

```python
word_class = []
def MecabNoun(text):
    m = mecab.Tagger ("-Ochasen")
    m.parse(' ')
    node = m.parseToNode(text)
    while node:
        if node.feature.split(",")[0] == "名詞":
            print(node.surface)
        node = node.next
    print("finish")
```

これは、文を``名詞``に分割してくれる``メソッド``である

```python
def MecabVerb(text):
    m = mecab.Tagger ("-Ochasen")
    m.parse(' ')
    node = m.parseToNode(text)
    while node:
        if node.feature.split(",")[0] == "動詞":
            print(node.surface)
        node = node.next
    print("finish")
```

これは、文を``動詞``に分割してくれる``メソッド``である

```python
def MecabAdjective(text):
    m = mecab.Tagger ("-Ochasen")
    m.parse(' ')
    node = m.parseToNode(text)
    while node:
        if node.feature.split(",")[0] == "形容詞":
            print(node.surface)
        node = node.next
    print("finish")
```
