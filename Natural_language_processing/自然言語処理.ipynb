```python
#単語頻出度カウント
def WordFrequencyCount(word):
        if word in wordFreq_dic:
            wordFreq_dic[word] +=1
        else:
            wordFreq_dic.setdefault(word, 1)
        return wordFreq_dic
```

```python
#特定の品詞の単語を抽出
mecab = MeCab.Tagger()
mecab.parse('')
for j in text_list:
    node = mecab.parseToNode(j)
    while node:
        if node.feature.split(",")[0] == "名詞":
            word = node.surface
            if word in value:
                WordFrequencyCount(word)
        else:
            pass
        node = node.next
```
